```python
# Make 'AGEONENTRY' ranges instead. Better for visualisation.
bins = [18, 21, 25, 30, 100]
labels = ['18-21','22-25','26-30','30+']
df['AGEGROUP'] = pd.cut(df['AGEONENTRY'],
                        bins=bins,
                        labels=labels,
                        include_lowest=True) # Includes the lowest boundary because by default it excludes 18.
# View the unique values.
df['AGEGROUP'].unique()
```

```python
# Drop variables with no relevance.
df.drop(['STUDENTYEARMODULE', 'SID','NUMHUS'], axis= 1, inplace= True)
```
